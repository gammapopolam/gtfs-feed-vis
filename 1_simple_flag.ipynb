### 1. Вводные функции

Существует много библиотек, нацеленных на обработку фида GTFS и его анализа, однако считаю, что достаточно имплементировать несколько функций, занимающихся "вытягиванием" данных из классов GTFS в формате табличной структуры Dataframe.

Сперва необходимо импортировать все таблицы в dataframe:

```python
import pandas as pd
folder='k26_static'

agency=pd.read_csv(folder+r'\agency.txt', sep=',')
calendar=pd.read_csv(folder+r'\calendar.txt', sep=',')
calendar_dates=pd.read_csv(folder+r'\calendar_dates.txt', sep=',')
routes=pd.read_csv(folder+r'\routes.txt', sep=',')
trips=pd.read_csv(folder+r'\trips.txt', sep=',')
stops=pd.read_csv(folder+r'\stops.txt', sep=',')
stop_times=pd.read_csv(folder+r'\stop_times.txt', sep=',')
shapes=pd.read_csv(folder+r'\shapes.txt', sep=',')
```

Вывести информацию по одной из остановок по идентификатору можно следующим образом:

```python
stop_id=21
stop=stops[stops.stop_id==stop_id]
stop
```

Чтобы вывести информацию по всем прибытиям на остановку в фиде, можно обратиться к классу stop_times.
Вместе с каждым прибытием на остановку можно сразу выводить информацию о поездке каждого из прибытий с помощью связи таблиц по идентификатору trip_id:

```python
st_selection=stop_times[stop_times.stop_id==stop_id]
st_selection_merge=st_selection.merge(trips, on='trip_id', how='left')
# пока что на данном этапе нет вывода информации по конкретному дню недели или делению на будний или выходной день
st_selection_merge
```

Только поездки: 

```python
trip_ids=list(set(st_selection.trip_id))
t_selection=trips[trips.trip_id.isin(trip_ids)]
t_selection
```

Какие маршруты проходят через остановку?

```python
route_ids=list(set(t_selection.route_id))
r_selection=routes[routes.route_id.isin(route_ids)]
r_selection
```

Теперь задача сложнее: нам нужно узнать первый и последний рейсы каждого из маршрутов, проходящих через остановку. На данном этапе - без учета по конкретным дням недели или делению на будний или выходной день

```python
route_ids=list(r_selection.route_id)
for route_id in route_ids:
    route=r_selection[r_selection.route_id==route_id]
    route_arrives=st_selection_merge[st_selection_merge.route_id==route_id]
    
    # Без учета календаря
    max=route_arrives.arrival_time.max()
    min=route_arrives.arrival_time.min()
    # Без учета возможных укоротов/вариантов
    main_headsign=route_arrives.trip_headsign.mode().item()
    print(f'Маршрут {route['route_short_name'].item()} до {main_headsign}: {min}-{max}')
```

Мы получили всё для того, чтобы получить данные для визуализации флага остановки. Обернем все в словарь и дополнительно укажем цвет маршрута и цвет текста маршрута:

```python
route_ids=list(r_selection.route_id)
route_info=dict()
for route_id in route_ids:
    route=r_selection[r_selection.route_id==route_id]
    route_arrives=st_selection_merge[st_selection_merge.route_id==route_id]
    
    # Без учета календаря
    max=route_arrives.arrival_time.max()
    min=route_arrives.arrival_time.min()
    # Без учета возможных укоротов/вариантов
    main_headsign=route_arrives.trip_headsign.mode().item()
    route_info[route_id]={'start': min, 'end': max, 'headsign': main_headsign,
                                  'route_color': route['route_color'].item(), 
                                  'route_text_color': route['route_text_color'].item(), 'route_short_name': route['route_short_name'].item()}
route_info
```

Вся последовательность действий обернута в класс Feed в скрипте feeds.py. 

Теперь можно приступить непосредственно к визуализации флага:

```python
from fpdf import FPDF

from feed import Feed

def HexToRGB(h):
    # Перевод HTML цвета в RGB
    h = h.lstrip('#')
    return tuple(int(h[i:i+2], 16) for i in (0, 2, 4))
def ServiceTime(start, end):
    # Исключение секунд в времени работы маршрута
    return f"{start.split(':')[0]}:{start.split(':')[1]}...{end.split(':')[0]}:{end.split(':')[1]}"

stop_id=21
feed=Feed('k26_static')
stop=feed.DescribeForFlag(stop_id)
stop_name=feed.StopInfo(stop_id).stop_name.item()
print(stop)
# Объявление класса PDF-файла
pdf = FPDF(format='A4')
pdf.add_font('MoscowSans', '', r"D:\fonts\Moscow Sans\MoscowSans-Regular.otf", uni=True)
pdf.alias_nb_pages()
# Добавляем страницу
pdf.add_page()
# Перед тем, как добавлять текстовый блок, необходимо задать параметры: шрифт, вес, кегль
pdf.set_font('MoscowSans', '', 40)
# Добавляем пиктограмму: координаты точки крепления x,y и размер пиктограммы w,h 
pdf.image("stop-01.png", x=10, y=10, w = 50, h = 50)
# Отступ от картинки для нового текстового блока
pdf.cell(60)
# Текстовый блок с автопереносом: размер блока, текст, обводка, центрирование
pdf.multi_cell(120, 30, stop_name, 0, 'L')
# Меняем размер шрифта
pdf.set_font('MoscowSans', '', 32)
# Переход на новую строку
pdf.ln(h=20)
pdf.cell(190, 20, 'Маршруты', 0, 1, 'L')
pdf.ln(h=5)
# Для каждого маршрута, проходящего через остановку, визуализируем информацию
for route in list(stop.keys()):
    pdf.set_font('MoscowSans', '', 40)
    # Объявляем цвет фона текстового блока и цвет текста в нем
    pdf.set_fill_color(*HexToRGB(stop[route]['route_color']))
    pdf.set_text_color(*HexToRGB(stop[route]['route_text_color']))
    pdf.cell(30, 20, str(stop[route]['route_short_name']), 0, 0, 'C', True)
    # Отступы
    pdf.cell(5, 10, '', 0, 0, 'L', False)
    # Черный цвет текста
    pdf.set_text_color(r=0)
    # Направление маршрута
    pdf.cell(160, 15, str(stop[route]['headsign']), 0, 0, 'L', False)
    pdf.ln(h=15)
    # Отступы
    pdf.cell(35, 20, '', 0, 0, 'L', False)
    pdf.set_font('MoscowSans', '', 24)
    # Время первого и последнего рейса
    pdf.cell(160, 5, ServiceTime(stop[route]['start'],stop[route]['end']), 0, 0, 'L', False)
    pdf.ln(h=15)

# Экспортируем результат
pdf.output(f'1_Extract.pdf')
```

Скрипт можно дополнять бесконечно, делая его приятнее человеческому глазу. Но главное - не нужно в векторном редакторе своими руками всё настраивать!
